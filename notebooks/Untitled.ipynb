```python
from PIL import Image
import urllib.parse
import shutil
import os

def reduce_image_quality(input_path, output_path, quality):
    try:
        # Open the image file
        image = Image.open(input_path)
        
        # Convert to RGB if the image has an alpha channel
        if image.mode == 'RGBA':
            image = image.convert('RGB')
        
        # Reduce the image quality
        image.save(output_path, format='JPEG', quality=quality, optimize=True)
        
        print(f"Image quality reduced and saved as: {output_path}")
        
    except IOError:
        print(f"Unable to open or process image file: {input_path}")

# Example usage
input_image_filename = 'apple.png'
input_image_path = os.path.join(os.getcwd(), input_image_filename)
output_image_path = 'output_image.jpg'
desired_quality = 50

reduce_image_quality(input_image_path, output_image_path, desired_quality)

# # Provide a download link for the output image in the Jupyter Notebook
# output_image_url = urllib.parse.quote(output_image_path)

# # Move the output image to a folder accessible by the Jupyter Notebook
# download_folder = './downloads/'
# shutil.move(output_image_path, download_folder)

# # Display the download link
# download_link = f'<a href="{download_folder}{output_image_url}" download>Download Reduced Image</a>'
# display(HTML(download_link))
```
