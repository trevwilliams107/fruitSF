```python
import pandas as pd
from IPython.display import display
import folium
```

```python
df = pd.read_csv('../data/Street_Tree_List.csv')
```

```python
loquats = df[df['qSpecies'].str.contains('loquat', case=False)]
# loquats = loquats[loquats['qSpecies'] != 'Eriobotrya deflexa :: Bronze Loquat']
# loquats = loquats.dropna(subset=['Latitude', 'Longitude'])
loquats.value_counts()
```

```python
species = loquats.groupby('qSpecies').first()
species
```

```python
# Create a map centered on sf
latitude_center = 37.7749
longitude_center = -122.4194
map_center = [latitude_center, longitude_center]
map_zoom = 12.5  # Adjust the zoom level as needed
map_obj = folium.Map(location=map_center, zoom_start=map_zoom)

# Add markers for each point
for index, row in loquats.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']
    marker = folium.Marker(location=[latitude, longitude])
    marker.add_to(map_obj)

# Display the map
map_obj
```
